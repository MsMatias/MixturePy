```python
seed(123)
def createBetas (X, a, b, n):
    ns = randint(a, b)
    r = np.random.uniform(1, 0.2, ns)
    rn = r/r.sum()
    id = sample(range(n), ns)
    betas = np.repeat(.0, n, axis=0)
    for index, i in enumerate(id, start = 0):
        betas[i] = rn[index]
        
    vector = X.to_numpy(copy=True)    
    vector = vector.flatten()
        
    A = np.dot(X, betas)
    AA = np.dot(X, betas) + vector[sample(range(len(vector)), len(X.index))].T
    
    return betas,id,A, AA

a = Parallel(n_jobs=4, backend='threading')(delayed(createBetas)(X = X, a = 2, b = 8, n = len(X.columns)) for i in range(1000))
```
