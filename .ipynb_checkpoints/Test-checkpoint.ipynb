```python
import numpy as np
import pandas as pd
import os, math

from sklearn.svm import NuSVR
from sklearn.preprocessing import StandardScaler
```

```python
# Read xlsx files
Y = pd.read_excel('BRCA.subsample.xlsx', sheet_name = 0)
X = pd.read_excel('LM22Signature.xlsx', sheet_name = 0)
```

```python
mean = X.iloc[:, 1:].mean()
std = X.iloc[:, 1:].stack().std()
X.iloc[:, 1:] = (X.iloc[:, 1:] - mean.mean()) / std
    
# Intersection between X and Y
X = X.loc[X['Gene symbol'].isin(Y['Gene symbol'])]
Y = Y.loc[Y['Gene symbol'].isin(X['Gene symbol'])]
    
# Ordering by GeneSymbol
X = X.sort_values(by=['Gene symbol'])
Y = Y.sort_values(by=['Gene symbol'])

# Slice Gene symbol column
X = X.iloc[:, 1:]
Y = Y.iloc[:, 1:]
X.reset_index(drop=True, inplace=True)
    
# Normalize features Matrix
scaler = StandardScaler()
scaler.fit(Y)
Yn = pd.DataFrame(scaler.transform(Y.values), columns=Y.columns, index=Y.index)
```

```python
wsel = [1 for x in range(len(X.columns))]
wsel = pd.DataFrame(wsel, X.columns).T
```

```python
# Run NuSVR
clf = NuSVR(kernel='linear', C=1.0, nu=0.25)
clf.fit(X, Y.iloc[:,1])
d = [[1, clf]]
listNuSvr = pd.DataFrame(d, columns=['RMSEpred', 'model'])
```

```python
model = listNuSvr.iloc[0]['model']

# Get betas
w = model.coef_

# Set values i to zero where i < 0
w = np.where(w<0, 0, w)
absW = w
        
# Normalize data
w = w/np.sum(w)
        
# Set values i to zero where i < delta
w = np.where(w<0.007, 0, w)

wa = pd.DataFrame(w[0], X.columns).T
wsel
```
